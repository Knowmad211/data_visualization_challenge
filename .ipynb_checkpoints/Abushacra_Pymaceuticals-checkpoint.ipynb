### Analysis

- The Regimen group sizes and sex distributions are relatively even. The Capomulin and Ramicane groups have slightly more mice in their groups, which would help get more accurate data. This is especially true when compared to the Propriva group, which has more than 30% fewer mice, and thus fewer data points. The distribution of male and female mice is also consistent across the groups which allows a proper test across sex.

- In the Summary Statistics table, Capomulin and Ramicane both have the lowest mean tumor volume, as well as the lowest standard deviation and error from the mean. This could indicate that these two regimens are the most effective treatment regimens, in addition to being the most consistent. Also, the Infubinol group has an outlier which would indicate a larger variability in treatment results. An analysis of tumor volume over time by each regimen would be helpful in determining the overall efficacy of these regimens. 

- After looking at the Correlation and Regression chart, here seems to be a positive correlation between mouse weight and tumor volume size within the Capomulin group, further supported by the .84 correlation coefficient between the two variables. This could mean a variety of things, however I feel as though it would be helpful if we could break down the data further to group mice with similar starting weights together. This could allow us to see whether or not this relatioinship is more complicated than just "bigger mouse = bigger tumor".
 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "Starter_Code/Pymaceuticals/data/Mouse_metadata.csv"
study_results_path = "Starter_Code/Pymaceuticals/data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
cmd = pd.merge(study_results, mouse_metadata, how="left", on=["Mouse ID", "Mouse ID"])

# Display the data table for preview
cmd.head()
```

```python
# Checking the number of mice.
mice_count = cmd["Mouse ID"].nunique()
mice_count
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicate_mice = cmd[cmd.duplicated(subset=["Mouse ID", "Timepoint"])]["Mouse ID"].unique()
duplicate_mice
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
cmd.loc[cmd["Mouse ID"]=="g989"]
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
ccmd = cmd.loc[cmd["Mouse ID"]!="g989"]
ccmd.head()
```

```python
# Checking the number of mice in the clean DataFrame.
ccmd["Mouse ID"].nunique()
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary DataFrame.



ccmd_means = ccmd.groupby("Drug Regimen")["Tumor Volume (mm3)"].mean()
ccmd_medians = ccmd.groupby("Drug Regimen")["Tumor Volume (mm3)"].median()
ccmd_variance = ccmd.groupby("Drug Regimen")["Tumor Volume (mm3)"].var()
ccmd_stddev = ccmd.groupby("Drug Regimen")["Tumor Volume (mm3)"].std()
ccmd_sem = ccmd.groupby("Drug Regimen")["Tumor Volume (mm3)"].sem()


summary_table = pd.DataFrame({
    "Mean Tumor Volume":ccmd_means,
    "Median Turmor Volume":ccmd_medians,
    "Tumor Volume Variance":ccmd_variance,
    "Tumor Volume Std. Dev.":ccmd_stddev,
    "Tumor Volume Std. Err.":ccmd_sem
})

summary_table
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)

# Using the aggregation method, produce the same summary statistics in a single line
advanced_summary_table = ccmd.groupby("Drug Regimen").agg({"Tumor Volume (mm3)":["mean","median","var","std","sem"]})
advanced_summary_table
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
max_timepoints = ccmd.groupby("Mouse ID")["Timepoint"].max()
final_tumor_volume = max_timepoints.reset_index()

# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
max_tumor_df = final_tumor_volume.merge(ccmd, how = "left", on = ["Mouse ID", "Timepoint"])
max_tumor_df
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatments = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]

# Create empty list to fill with tumor vol data (for plotting)
tumor_volume_data = []

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
for treatment in treatments:
    
    # Locate the rows which contain mice on each drug and get the tumor volumes
    tumor_volume = max_tumor_df.loc[max_tumor_df["Drug Regimen"] == treatment, "Tumor Volume (mm3)"]
    
    # add subset 
    tumor_volume_data.append(tumor_volume)

    #IQR Calculations
    quartiles = tumor_volume.quantile([0.25, 0.5, 0.75])
    lower_q = quartiles[0.25]
    upper_q = quartiles[0.75]
    iqr = upper_q - lower_q

    # Determine outliers using upper and lower bounds
    lower_bound = lower_q - (1.5*iqr)
    upper_bound = upper_q + (1.5*iqr)

    outliers = tumor_volume[(tumor_volume < lower_bound) | (tumor_volume > upper_bound)]

    print(f"{treatment}'s potential outliers: {outliers}")
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.
plt.figure(figsize=(7, 5))
plt.boxplot(tumor_volume_data, labels=treatments, sym="ro")
plt.xlabel("Drug Regimen")
plt.ylabel("Final Tumor Volume (mm3)")
plt.show()
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a single mouse treated with Capomulin
# Select Mouse (1st)
chosen_mouse_id = ccmd.loc[ccmd["Drug Regimen"] == "Capomulin", "Mouse ID"].iloc[0]
chosen_mouse_data = ccmd.loc[(ccmd["Mouse ID"] == chosen_mouse_id) & (ccmd["Drug Regimen"] == "Capomulin")]

# Plot
plt.figure(figsize=(7, 5))
plt.plot(chosen_mouse_data["Timepoint"], chosen_mouse_data["Tumor Volume (mm3)"])
plt.title(f"Capomulin Treatment of Mouse {chosen_mouse_id}")
plt.xlabel("Timepoint (days)")
plt.ylabel("Tumor Volume (mm3)")
plt.show()
```

```python
# Generate a scatter plot of mouse weight vs. the average observed tumor volume for the entire Capomulin regimen
# Create table with Average Tumor Volume and weight
capomulin_atv = ccmd[ccmd["Drug Regimen"] == "Capomulin"].groupby("Mouse ID")["Tumor Volume (mm3)"].mean()
capomulin_data = pd.merge(capomulin_atv, mouse_metadata, on="Mouse ID")
print(capomulin_data)
```

```python
# Plot
plt.figure(figsize=(7, 5))
plt.scatter(capomulin_data["Weight (g)"], capomulin_data["Tumor Volume (mm3)"])
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model for mouse weight and average observed tumor volume for the entire Capomulin regimen
correlation_coefficient = st.pearsonr(capomulin_data["Weight (g)"], capomulin_data["Tumor Volume (mm3)"])[0]
print(f"The correlation coefficient between mouse weight and average tumor volume is {correlation_coefficient:.2f}")

(slope, intercept, rvalue, pvalue, stdderr) = st.linregress(capomulin_data["Weight (g)"], capomulin_data["Tumor Volume (mm3)"])

plt.figure(figsize=(7,5))
plt.scatter(capomulin_data["Weight (g)"], capomulin_data["Tumor Volume (mm3)"])

regression_values = slope * capomulin_data["Weight (g)"] + intercept
plt.plot(capomulin_data["Weight (g)"], regression_values, "r-")


plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")

plt.show()
```
